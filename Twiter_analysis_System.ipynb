# Filling null value via median and mode

```
1) As our datasets contains maximum outliers so mean is affected via outliers so median is suitaible for the datasets
2) for categorical datasets has some na value for that we will use mode
3) In our datasets some cololum contains false and 'False' value to made our datasets consistant  we replace the value
```

````python
df4['quote_count']=df4['quote_count'].fillna(df4['quote_count'].median())
df4['like_count']=df4['like_count'].fillna(df4['like_count'].median())
df4['reply_count']=df4['reply_count'].fillna(df4['reply_count'].median())
df4['retweet_count']=df4['retweet_count'].fillna(df4['retweet_count'].median())
df4['author_follower_count']=df4['author_follower_count'].fillna(df4['author_follower_count'].median())
df4['source']=df4['source'].fillna(df4['source'].mode()).astype(str)
df4['reply_settings']=df4['reply_settings'].fillna(df4['reply_settings'].mode()).astype(str)
df4['possibly_sensitive']=df4['possibly_sensitive'].fillna(df4['possibly_sensitive'].mode()).astype(str)
df4['lang']=df4['lang'].fillna(df4['lang'].mode()).astype(str)
df4['created_at']=df4['created_at'].fillna(df4['created_at'].mode()).astype(str)
df4['author_handle']=df4['author_handle'].fillna(df4['author_handle'].mode()).astype(str)
df4['text']=df4['text'].fillna(df4['text'].mode())
````

````python
df4['created_at']=pd.to_datetime(df4['created_at'],utc=True)
````

## Remove 'False' string litral from data sets
In our datasets some cololum contains False and 'False',True and 'True' value to made our datasets consistant  we replace the value

````python
import ast
````

````python
def clean_boolean_data(df):
    bool_data=[]
    for x in df:
        bool_data.append(ast.literal_eval(x.strip().title()))
    return bool_data
````

````python
df4['from_stream']=clean_boolean_data(df4['from_stream'].astype(str))
df4['directly_from_stream']=clean_boolean_data(df4['directly_from_stream'].astype(str))
df4['directly_from_search']=clean_boolean_data(df4['directly_from_search'].astype(str))
df4['from_quote_search']=clean_boolean_data(df4['from_quote_search'].astype(str))
df4['directly_from_quote_search']=clean_boolean_data(df4['directly_from_quote_search'].astype(str))
df4['directly_from_convo_search']=clean_boolean_data(df4['directly_from_convo_search'].astype(str))
df4['from_convo_search']=clean_boolean_data(df4['from_convo_search'].astype(str))
df4['from_timeline_search']=clean_boolean_data(df4['from_timeline_search'].astype(str))
df4['directly_from_timeline_search']=clean_boolean_data(df4['directly_from_timeline_search'].astype(str))
````

## Analysis Part from Data Sets


````python
search_term = str(input())

required_df=df4[df4['text'].str.contains(search_term, case=False, na=False)]
rows, cols =required_df.shape
print(f'Total Unique Tweet Posted containing {search_term}',required_df['id'].nunique())
print(f'Total Tweet Posted containing {search_term} Each day \n', required_df['created_at_Weekday'].value_counts().reset_index().rename(columns={'count':'Tweets'}))
print(f'Times of day were the tweets posted on {search_term} \n', required_df['time_of_day'].value_counts().reset_index().rename(columns={'count':'Tweets'}))
print(f'The user posted tweets on {search_term} \n',required_df['author_handle'].value_counts().reset_index().rename(columns={'count':'Tweets'}))
print(f'The Average Likes on {search_term} = ',required_df['like_count'].mean())
print(f'Source from {search_term} Post \n',required_df['source'].value_counts().reset_index().rename(columns={'count':'Tweets'}))
````

````python
df4['created_at_hour'] = pd.to_datetime(df4['created_at']).dt.hour
df4['created_at_second']=pd.to_datetime(df4['created_at']).dt.second
df4['created_at_min']=pd.to_datetime(df4['created_at']).dt.minute
df4['created_at_day'] = pd.to_datetime(df4['created_at']).dt.day
df4['created_at_month'] = pd.to_datetime(df4['created_at']).dt.month
df4['created_at_year']=pd.to_datetime(df4['created_at']).dt.year
df4['created_at_Weekday']=pd.to_datetime(df4['created_at']).dt.weekday
df4["created_at_Weekday"] = df4["created_at_Weekday"].map({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
df4.loc[(df4.created_at_hour >= 6) & (df4.created_at_hour < 12) , 'time_of_day'] = 'Morning'
df4.loc[(df4.created_at_hour >= 12) & (df4.created_at_hour < 15) , 'time_of_day'] = 'Afternoon'
df4.loc[(df4.created_at_hour >= 15) & (df4.created_at_hour < 18) , 'time_of_day'] = 'Evening'
df4.loc[(df4.created_at_hour >= 18) | (df4.created_at_hour < 6) , 'time_of_day'] = 'Night'

df4.drop('created_at',axis=1,inplace=True)
````
